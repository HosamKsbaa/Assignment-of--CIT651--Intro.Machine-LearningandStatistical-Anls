Prove this statement is true using Exponentially distributed population
    

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#assuming
rate = 0.25
#Population mean is going to be
mu = 1/rate

print(3.3%3)

```

```python
# def simulate_central_limit(random_samples):
#     '''
#     This function simulates the central limit theorem by simulating rolling a dice for n_rolls times and averaging the outcomes.
#     Input:The number of dice which we throw.
#     Output: histogram of the average outcomes, a vector with the average of each roll, 
#     and the mean and variance of the total average
#     outcomes.
#     '''
def simulate_central_limit(random_samples,sample_size):
   
    samples = np.zeros((random_samples, sample_size))  

    for i in range(random_samples):

        samples[i] = np.random.exponential(mu, sample_size) 
    sample_means = samples.mean(axis=1)  
    fig, ax = plt.subplots()
    ax.hist(sample_means, bins=30,range=(2, 6))

    plt.show()
    
    mean_of_means = sample_means.mean()
    var_of_means = sample_means.var()
    
    return sample_means, mean_of_means, var_of_means

```

```python
#First exp
# start with 50 random samples, sample size = 2 

avgs1, mean1, var1 = simulate_central_limit(500,2)

avgs3, mean3, var3 = simulate_central_limit(500,50)

avgs3, mean3, var3 = simulate_central_limit(500,100)

#Second exp
# use another 50 random samples ,sample size= 500
avgs2, mean2, var2 = simulate_central_limit(500,500)

print("First 50 random samples: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean1, var1))
print("Average of outcomes in each case: ", avgs1, "\n")

print("Another 50 random samples: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean2, var2))
print("Average of outcomes in each case: ", avgs2, "\n")
```
